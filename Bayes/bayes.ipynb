## Problem



- $\prob{\pos\given\sick} = \alpha$
- $ \prob{ \pos\given\heal } = \beta $
- $\prob{\sick} = \gamma$

Find $\prob{\heal\given\pos}$

Bayes theorem: $$\prob{\heal\given\pos}  = \prob{\pos\given\heal} \frac{\prob{\heal}}{\prob{\pos}}$$

Denominator by Total probability theorem:

$\prob{\pos} = \prob{\pos \and \heal} + \prob{\pos \and \sick} = \prob{\pos \given \heal} \prob{\heal} + \prob{\pos\given\sick}{\prob{\sick}}$

Putting togehter: 

$$\prob{\heal\given\pos}  =  \frac{\prob{\pos\given\heal} \prob{\heal}}{\prob{\pos \given \heal} \prob{\heal} + \prob{\pos\given\sick}{\prob{\sick}}} =  \frac{\beta (1-\gamma)}{\beta(1-\gamma) + \alpha\gamma}$$

```python
alpha = 99/100
beta = 2/100
gamma = 1e-3
```

```python
Php = beta * (1-gamma) / (beta * (1-gamma) + alpha * gamma )
Php
```

## Problem
Assume now a test is performed and gives a positive result $n$ times. What is the prob to be healty as a function of $n$?

$$\prob{\heal\given\pos^n}  =
\frac{ \prob{\pos\given\heal}^n \prob{\heal} } {\prob{\pos \given \heal}^n \prob{\heal} + \prob{\pos\given\sick}^n{\prob{\sick}}}
=  \frac{\beta^n (1-\gamma)}{ \beta^n(1-\gamma) + \alpha^n\gamma}$$

```python
def Phpn(n):
    return beta**n * (1-gamma) / ( beta**n * (1-gamma) + alpha**n * gamma )
```

```python
Phpn(1)
```

```python
Phpn(2)
```

```python
Phpn(3)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(10)
plt.plot(x, Phpn(x), 'o--')
plt.xlabel("Number of tests")
plt.ylabel("Probability to be heatly given n positive tests")
plt.show()
```
